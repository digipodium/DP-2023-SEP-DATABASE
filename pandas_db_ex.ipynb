Fetching data from mysql database using pandas

```python
host = 'localhost'
port = 3306
username = 'root'
password = 'root'
database = 'sales'
protocol = 'mysql://'
conn = f'{protocol}{username}:{password}@{host}:{port}/{database}'
print(conn)
```

```python
query = '''SELECT 
    COUNT(CID) as total,
    state,
    if(count(cid) >= 10,"✅", "❌") as importance
FROM customers
GROUP BY state
ORDER BY importance;
'''
df_result = pd.read_sql(query, con=conn)
df_result.total.plot(style='o-', figsize=(10, 5))
```
