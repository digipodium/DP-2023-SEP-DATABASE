#  count the number of movies on the basis of rental\_duration and rating

```sql
-- ye h comment
select 
    count(film_id) as number_of_films, 
    rental_duration,
    rating
from film
group by rating, rental_duration
order by rental_duration, rating;
```

```sql
select 
    MAX(replacement_cost),
    MIN(replacement_cost),
    AVG(replacement_cost),
    rating
from film
group by rating;
```

```sql
select 
    count(film_id) as number_of_films, 
    rental_duration,
    rating
from film
group by rating, rental_duration
order by rental_duration, rating;
```

```sql
select 
    count(film_id) as number_of_films, 
    rental_duration,
    rating
from film
group by 
    rating, rental_duration
having 
    number_of_films > 40
order by 
    rental_duration, rating;
```
