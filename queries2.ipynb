SELECT query with **where** clause

```sql
SELECT 
    first_name,
    last_name,
    store_id
FROM
    customer
WHERE
    store_id = 1;
```

```sql
-- Counting Inactive Customers
SELECT 
    COUNT(first_name) AS inactive_customers
FROM
    customer
WHERE 
    active = 0;
```

```sql
SELECT
    first_name,
    store_id
FROM
    customer
WHERE 
    store_id BETWEEN 1 AND 2;
```

```sql
SELECT * FROM city;
```

```sql
SELECT 
    city,
    country_id
FROM
    city
WHERE 
    country_id IN (15, 6, 38);
```

```sql
SELECT * FROM film;
```

```sql
SELECT 
    title,
    rating 
FROM 
    film
ORDER BY
    rating;
```

```sql
SELECT
    DISTINCT rating
FROM
    film
```

```sql
SELECT 
    title,
    rating
FROM
    film
ORDER BY FIELD(rating, 'PG-13', 'PG', 'G', 'NC-17', 'R')
```

```sql
SELECT 
    title, length, rating
FROM
    film
WHERE 
    length <= 100
ORDER BY 
    length DESC;
```

```sql
SELECT * from actor;
```

```sql
-- % mean 0 or more characters
-- _ mean 1 character
SELECT 
    first_name,
    last_name
FROM 
    actor
WHERE 
    last_name LIKE '%a%';
```

```sql
SELECT 
    title,
    length,
    rating,
    rental_duration
FROM
    film
WHERE length < 100
AND rating IN ('PG-13', 'PG')
AND rental_duration < 5
ORDER BY 
    length DESC;
```
